```python
def foo(g):
    min1=[]
    max1=[]
    for i in g:
        q1,q3=g[i].quantile([.25,.75])
        min=q1-(q3-q1)*1.5
        max=q3+(q3-q1)*1.5
        min1.append(min)
        max1.append(max)
    return min1,max1
```

```python
h=foo(y1)
```
